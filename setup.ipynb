```python
import pybullet as p
import pybullet_data as pd

def make_arena(arena_size=10, wall_height=1,pid = 0):
    wall_thickness = 0.5
    floor_collision_shape = p.createCollisionShape(shapeType=p.GEOM_BOX, halfExtents=[arena_size/2, arena_size/2, wall_thickness], physicsClientId=pid)
    floor_visual_shape = p.createVisualShape(shapeType=p.GEOM_BOX, halfExtents=[arena_size/2, arena_size/2, wall_thickness], rgbaColor=[1, 1, 0, 1], physicsClientId=pid)
    floor_body = p.createMultiBody(baseMass=0, baseCollisionShapeIndex=floor_collision_shape, baseVisualShapeIndex=floor_visual_shape, basePosition=[0, 0, -wall_thickness], physicsClientId=pid)

    wall_collision_shape = p.createCollisionShape(shapeType=p.GEOM_BOX, halfExtents=[arena_size/2, wall_thickness/2, wall_height/2], physicsClientId=pid)
    wall_visual_shape = p.createVisualShape(shapeType=p.GEOM_BOX, halfExtents=[arena_size/2, wall_thickness/2, wall_height/2], rgbaColor=[0.7, 0.7, 0.7, 1], physicsClientId=pid)  # Gray walls

        # Create four walls
    p.createMultiBody(baseMass=0, baseCollisionShapeIndex=wall_collision_shape, baseVisualShapeIndex=wall_visual_shape, basePosition=[0, arena_size/2, wall_height/2], physicsClientId=pid)
    p.createMultiBody(baseMass=0, baseCollisionShapeIndex=wall_collision_shape, baseVisualShapeIndex=wall_visual_shape, basePosition=[0, -arena_size/2, wall_height/2], physicsClientId=pid)

    wall_collision_shape = p.createCollisionShape(shapeType=p.GEOM_BOX, halfExtents=[wall_thickness/2, arena_size/2, wall_height/2], physicsClientId=pid)
    wall_visual_shape = p.createVisualShape(shapeType=p.GEOM_BOX, halfExtents=[wall_thickness/2, arena_size/2, wall_height/2], rgbaColor=[0.7, 0.7, 0.7, 1], physicsClientId=pid)  # Gray walls

    p.createMultiBody(baseMass=0, baseCollisionShapeIndex=wall_collision_shape, baseVisualShapeIndex=wall_visual_shape, basePosition=[arena_size/2, 0, wall_height/2], physicsClientId=pid)
    p.createMultiBody(baseMass=0, baseCollisionShapeIndex=wall_collision_shape, baseVisualShapeIndex=wall_visual_shape, basePosition=[-arena_size/2, 0, wall_height/2], physicsClientId=pid)
```

```python
import pygetwindow as gw
import genomeprelimbssymmetry
import creatureprereactivemotor
def main2(csv_file):
    assert os.path.exists(csv_file), "Tried to load " + csv_file + " but it does not exists"

    physicsClientId = p.connect(p.GUI)
    p.setPhysicsEngineParameter(enableFileCaching=0, physicsClientId=physicsClientId)
    p.configureDebugVisualizer(p.COV_ENABLE_GUI, 0)
    p.setGravity(0, 0, -10)

    # Common PyBullet window titles (may vary slightly)
    possible_titles = ["PyBullet Physics Server", "bullet.py", "PyBullet", "pybullet"] 
    pb_window = None
    for title in possible_titles:
        windows = gw.getWindowsWithTitle(title)
        if windows:
            pb_window = windows[0]
            break

    if pb_window:
        try:
            # For Windows/Linux: activate() or minimize()/maximize()
            pb_window.activate()
            # On some systems, you might need to try maximize or restore after activate
            # pb_window.restore() # Ensure it's not minimized
            # pb_window.maximize() # If you want it full screen
        except Exception as e:
            print(f"Could not activate PyBullet window: {e}")
            print("You may need to manually click on the PyBullet window in your taskbar.")
    else:
        print("PyBullet window not found, could not bring to foreground.")
    # --- End window focus attempt ---

    arena_size = 23
    make_arena(arena_size=arena_size, pid=physicsClientId)

    mountain_position = (0, 0, -1)
    mountain_orientation = p.getQuaternionFromEuler((0, 0, 0))
    p.setAdditionalSearchPath('shapes/')
    mountain_id = p.loadURDF("gaussian_pyramid_larger_wider.urdf", mountain_position, mountain_orientation, useFixedBase=1,physicsClientId=physicsClientId)
    
    # !!! IMPORTANT: Define the mountain peak height used during training !!!
    # Ensure this matches the `guassian_pyramid_height` in your simulation.py::run_creature
    MOUNTAIN_PEAK_HEIGHT = 6 
    # --- SET CAMERA POSITION ---
    # Aim the camera at the approximate center of the mountain peak (0,0,MOUNTAIN_PEAK_HEIGHT)
    # Adjust cameraDistance, cameraYaw, cameraPitch as needed for desired view
    camera_distance = 15 # Adjust distance from target
    camera_yaw = 45      # Angle around the Z-axis (0-360)
    camera_pitch = -60   # Angle up/down (-90 (straight down) to 90 (straight up))
    camera_target_position = [0, 0, 0]

    p.resetDebugVisualizerCamera(camera_distance, camera_yaw, camera_pitch, camera_target_position)
    # --- END SET CAMERA POSITION ---

    cr = creatureprereactivemotor.Creature(gene_count=2) # Gene count doesn't matter here as DNA is loaded
    dna = genomeprelimbssymmetry.Genome.from_csv(csv_file)
    cr.update_dna(dna) # This resets cr.start_position, last_position, closest_position to None

    # Save and load the URDF (as you had it)
    with open('test.urdf', 'w') as f:
        f.write(cr.to_xml())
    rob1 = p.loadURDF('test.urdf')

    # Disable self-collision (as you had it)
    num_joints = p.getNumJoints(rob1)
    for i in range(-1, num_joints):
        for j in range(i + 1, num_joints):
            if i == j: continue
            p.setCollisionFilterPair(rob1, rob1, i, j, 0)

    # Air drop it to the starting position
    p.resetBasePositionAndOrientation(rob1, [6.5, 6.5, 8], [0, 0, 0, 1])

    # --- START OF SETTLEMENT PERIOD (CRITICAL) ---
    touched_ground = False
    exit_iter = 0
    while not touched_ground and exit_iter < (6 * 240): # Max 6 seconds to settle
        p.stepSimulation()
        contacts = p.getContactPoints(bodyA=rob1)
        for contact in contacts:
            if contact[2] == mountain_id: # bodyB is the mountain_id (or 0 for floor)
                touched_ground = True
                # Get initial position after settling
                start_pos, orn = p.getBasePositionAndOrientation(rob1)
                cr.update_position(start_pos) # Sets cr.start_position
                cr.closest_position = start_pos # Initialize closest_position after landing
                break
            # Add check for floor if mountain has a flat base, or if creature can land on arena floor
            # e.g., if contact[2] == floor_body_id (if you make the floor multiBody)
            # or if using a simple plane, contact[2] might be 0
            if contact[2] == 0 and p.getBodyInfo(contact[2])[0].decode('utf-8') == "plane":
                touched_ground = True
                start_pos, orn = p.getBasePositionAndOrientation(rob1)
                cr.update_position(start_pos) # Sets cr.start_position
                cr.closest_position = start_pos # Initialize closest_position after landing
                break

        exit_iter += 1
        # Optional: Reset creature if it takes too long to settle (from your training code)
        if exit_iter >= 3 * 240 and exit_iter <= ((3 * 240)+3):
            p.resetBasePositionAndOrientation(rob1, [7, 7, 8], [0, 0, 0, 1])

    if not touched_ground:
        print(f"Creature failed to settle after multiple attempts. Assigning poor fitness.")
        # Assign poor fitness if it never settled, just like in training
        cr.closest_position = [0,0,-999] 
        p.disconnect()
        return

    # Optional: give 1 sec (240 frames) for stability before motors
    for _ in range(240):
        p.stepSimulation()

    # --- END OF SETTLEMENT PERIOD ---

    # Main simulation loop
    elapsed_time = 0
    wait_time = 1.0 / 240  # seconds
    total_time = 30  # seconds
    step = 0
    final_distance_to_peak = 999.0 # Initialize with a bad value

    while True:
        p.stepSimulation()
        step += 1

        if step % 24 == 0:
            motors = cr.get_motors()
            # The assert check should now pass if creature and PyBullet are aligned
            assert len(motors) == p.getNumJoints(rob1), f"Something went wrong! Motors: {len(motors)}, Joints: {p.getNumJoints(rob1)}"
            for jid in range(p.getNumJoints(rob1)):
                mode = p.VELOCITY_CONTROL
                vel = motors[jid].get_output()
                p.setJointMotorControl2(rob1, 
                                        jid,  
                                        controlMode=mode, 
                                        targetVelocity=vel)
            
            # --- UPDATE CLOSEST_POSITION AND PRINT METRIC (CRITICAL) ---
            current_pos, _ = p.getBasePositionAndOrientation(rob1)
            cr.update_position(current_pos) # Updates last_position
            
            # Update closest_position if current position is closer to peak
            dist_from_peak_current = np.linalg.norm(np.asarray([0, 0, MOUNTAIN_PEAK_HEIGHT]) - np.asarray(current_pos))
            if cr.closest_position is None or dist_from_peak_current < np.linalg.norm(np.asarray([0, 0, MOUNTAIN_PEAK_HEIGHT]) - np.asarray(cr.closest_position)) :
                 cr.closest_position = current_pos

            # Print the actual metric used for training
            final_distance_to_peak = cr.distance_to_peak(height=MOUNTAIN_PEAK_HEIGHT)
            print(f"Current closest distance to peak: {final_distance_to_peak:.3f}")
            # --- END CRITICAL UPDATES ---

        time.sleep(wait_time) # Keep this for GUI playback
        elapsed_time += wait_time
        if elapsed_time > total_time:
            break

    print(f"\nFINAL CLOSEST DISTANCE TO PEAK: {final_distance_to_peak:.3f} meters")
    p.disconnect()
```

```python
import pygetwindow as gw
import genome
import creature
def update_motors( cid, cr,touch_data):
        """
        cid is the id in the physics engine
        cr is a creature object
        """
        motors = cr.get_motors() # Get the list of motors
        # Loop through all the JOINTS of the creature
        for jid in range(p.getNumJoints(cid)):
            # A joint connects a parent link and a child link.
            # The joint's index (jid) corresponds to the child link's index.
            # PyBullet's link indices start at 0, matching the joint indices.
            child_link_index = jid 

            # Check if this specific limb has touch data.
            # .get(key, False) is a safe way to check a dictionary.
            # It returns the value for the key if it exists, otherwise it returns False.
            is_touching = touch_data.get(child_link_index, False)
            
            m = motors[jid]

            # Call the motor's get_output method, now with the sensory information.
            vel = m.get_output(touch=is_touching)

            p.setJointMotorControl2(cid, jid, 
                    controlMode=p.VELOCITY_CONTROL, 
                    targetVelocity=vel, 
                    force = 10)
            
def main3(csv_file):
    assert os.path.exists(csv_file), "Tried to load " + csv_file + " but it does not exists"

    physicsClientId = p.connect(p.GUI)
    p.setPhysicsEngineParameter(enableFileCaching=0, physicsClientId=physicsClientId)
    p.configureDebugVisualizer(p.COV_ENABLE_GUI, 0)
    p.setGravity(0, 0, -10)

    # Common PyBullet window titles (may vary slightly)
    possible_titles = ["PyBullet Physics Server", "bullet.py", "PyBullet", "pybullet"] 
    pb_window = None
    for title in possible_titles:
        windows = gw.getWindowsWithTitle(title)
        if windows:
            pb_window = windows[0]
            break

    if pb_window:
        try:
            # For Windows/Linux: activate() or minimize()/maximize()
            pb_window.activate()
            # On some systems, you might need to try maximize or restore after activate
            # pb_window.restore() # Ensure it's not minimized
            # pb_window.maximize() # If you want it full screen
        except Exception as e:
            print(f"Could not activate PyBullet window: {e}")
            print("You may need to manually click on the PyBullet window in your taskbar.")
    else:
        print("PyBullet window not found, could not bring to foreground.")
    # --- End window focus attempt ---

    arena_size = 23
    make_arena(arena_size=arena_size, pid=physicsClientId)

    mountain_position = (0, 0, -1)
    mountain_orientation = p.getQuaternionFromEuler((0, 0, 0))
    p.setAdditionalSearchPath('shapes/')
    mountain_id = p.loadURDF("gaussian_pyramid_larger_wider.urdf", mountain_position, mountain_orientation, useFixedBase=1,physicsClientId=physicsClientId)
    
    # !!! IMPORTANT: Define the mountain peak height used during training !!!
    # Ensure this matches the `guassian_pyramid_height` in your simulation.py::run_creature
    MOUNTAIN_PEAK_HEIGHT = 6 
    # --- SET CAMERA POSITION ---
    # Aim the camera at the approximate center of the mountain peak (0,0,MOUNTAIN_PEAK_HEIGHT)
    # Adjust cameraDistance, cameraYaw, cameraPitch as needed for desired view
    camera_distance = 15 # Adjust distance from target
    camera_yaw = 45      # Angle around the Z-axis (0-360)
    camera_pitch = -60   # Angle up/down (-90 (straight down) to 90 (straight up))
    camera_target_position = [7, 7, 0]

    p.resetDebugVisualizerCamera(camera_distance, camera_yaw, camera_pitch, camera_target_position)
    # --- END SET CAMERA POSITION ---

    cr = creature.Creature(gene_count=2) # Gene count doesn't matter here as DNA is loaded
    dna = genome.Genome.from_csv(csv_file)
    cr.update_dna(dna) # This resets cr.start_position, last_position, closest_position to None

    # Save and load the URDF (as you had it)
    with open('test.urdf', 'w') as f:
        f.write(cr.to_xml())
    rob1 = p.loadURDF('test.urdf')

    # Disable self-collision (as you had it)
    num_joints = p.getNumJoints(rob1)
    for i in range(-1, num_joints):
        for j in range(i + 1, num_joints):
            if i == j: continue
            p.setCollisionFilterPair(rob1, rob1, i, j, 0)

    # Air drop it to the starting position
    p.resetBasePositionAndOrientation(rob1, [7, 7, 8], [0, 0, 0, 1])

    # --- START OF SETTLEMENT PERIOD (CRITICAL) ---
    touched_ground = False
    exit_iter = 0
    while not touched_ground and exit_iter < (6 * 240): # Max 6 seconds to settle
        p.stepSimulation()
        contacts = p.getContactPoints(bodyA=rob1)
        for contact in contacts:
            if contact[2] == mountain_id: # bodyB is the mountain_id (or 0 for floor)
                touched_ground = True
                # Get initial position after settling
                start_pos, orn = p.getBasePositionAndOrientation(rob1)
                cr.update_position(start_pos) # Sets cr.start_position
                cr.closest_position = start_pos # Initialize closest_position after landing
                break
            # Add check for floor if mountain has a flat base, or if creature can land on arena floor
            # e.g., if contact[2] == floor_body_id (if you make the floor multiBody)
            # or if using a simple plane, contact[2] might be 0
            if contact[2] == 0 and p.getBodyInfo(contact[2])[0].decode('utf-8') == "plane":
                touched_ground = True
                start_pos, orn = p.getBasePositionAndOrientation(rob1)
                cr.update_position(start_pos) # Sets cr.start_position
                cr.closest_position = start_pos # Initialize closest_position after landing
                break

        exit_iter += 1
        # Optional: Reset creature if it takes too long to settle (from your training code)
        if exit_iter >= 3 * 240 and exit_iter <= ((3 * 240)+3):
            p.resetBasePositionAndOrientation(rob1, [7, 7, 8], [0, 0, 0, 1])

    if not touched_ground:
        print(f"Creature failed to settle after multiple attempts. Assigning poor fitness.")
        # Assign poor fitness if it never settled, just like in training
        cr.closest_position = [0,0,-999] 
        p.disconnect()
        return

    # Optional: give 1 sec (240 frames) for stability before motors
    for _ in range(240):
        p.stepSimulation()

    # --- END OF SETTLEMENT PERIOD ---

    # Main simulation loop
    elapsed_time = 0
    wait_time = 1.0 / 240  # seconds
    total_time = 120  # seconds
    step = 0
    final_distance_to_peak = 999.0 # Initialize with a bad value

    while True:
        p.stepSimulation()
        step += 1

        if step % 24 == 0:
            
            touch_data = {} 
            contacts = p.getContactPoints(bodyA=rob1)
            for contact in contacts:
                # contact[3] is the linkIndex of bodyA (our creature) that made contact.
                limb_index = contact[3]
                touch_data[limb_index] = True

            if (step % 24 == 0) and touched_ground == True:
                update_motors(cid=rob1, cr=cr, touch_data = touch_data)
            
            # --- UPDATE CLOSEST_POSITION AND PRINT METRIC (CRITICAL) ---
            current_pos, _ = p.getBasePositionAndOrientation(rob1)
            cr.update_position(current_pos) # Updates last_position
            
            # Update closest_position if current position is closer to peak
            dist_from_peak_current = np.linalg.norm(np.asarray([0, 0, MOUNTAIN_PEAK_HEIGHT]) - np.asarray(current_pos))
            if cr.closest_position is None or dist_from_peak_current < np.linalg.norm(np.asarray([0, 0, MOUNTAIN_PEAK_HEIGHT]) - np.asarray(cr.closest_position)) :
                 cr.closest_position = current_pos

            # Print the actual metric used for training
            final_distance_to_peak = cr.distance_to_peak(height=MOUNTAIN_PEAK_HEIGHT)
            print(f"Current closest distance to peak: {final_distance_to_peak:.3f}")
            # --- END CRITICAL UPDATES ---

        time.sleep(wait_time) # Keep this for GUI playback
        elapsed_time += wait_time
        if elapsed_time > total_time:
            break

    print(f"\nFINAL CLOSEST DISTANCE TO PEAK: {final_distance_to_peak:.3f} meters")
    p.disconnect()       
```
